```python
n = int(input("enter the range number:"))
sum = 0
for i in range(0,n+1):
    sum = sum + i
print("Sum of n number :",sum)
```

```python
a = int(input("enter the number :"))
def fact(a):
    if a == 1 or a == 0:
        return 1
    else:
        return a*fact(a-1)
    
print("factorial of number",fact(a))
```
